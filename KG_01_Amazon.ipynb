```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data available at: https://www.kaggle.com/sootersaalu/exploratory-data-analysis
df = pd.read_csv("bestsellers with categories.csv")
df.head()
```

```python
#To overwrite the object?
data_order = df.groupby(['Name', 'Author'], as_index = False)[['User Rating','Reviews']].mean()
data_order = data_order[data_order['Reviews'] > 3000]
data_order = data_order.sort_values('User Rating', ascending=False)
data_order.head(20)
```

```python
#Weighted Rating
m = min(df['Reviews'])
C = df['User Rating'].mean()

def weighted_rating(x, m=m, C = C):
    v = x['Reviews']
    R = x['User Rating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
df['Weighted Rating'] = df.apply(weighted_rating, axis = 1)
```

```python

df.groupby(['Name', "Author", 'Genre'], as_index = False)[['User Rating', 'Reviews', 'Weighted Rating']].mean().sort_values(by = 'Weighted Rating', ascending = False).head(10)
```
